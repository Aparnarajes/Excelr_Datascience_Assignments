```python
# Program to check whether a number is prime or not
def is_prime(n):
    if n <= 1:
        return False
    if n == 2:
        return True
    if n % 2 == 0:
        return False

    for i in range(3, int(n ** 0.5) + 1, 2):
        if n % i == 0:
            return False

    return True


n = int(input("Enter a number: "))
print("Prime number" if is_prime(n) else "Not a prime number")
```

```python

#Product of Random Numbers
import random

num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

user_answer = num1 * num2  # assume user's answer

if user_answer == num1 * num2:
    print("Correct! Well done")
else:
    print("Incorrect")

print("Numbers were:", num1, "and", num2)
```

```python
# Printing squares of EVEN numbers between 100 and 200

for num in range(100, 201):
    if num % 2 == 0:   # checking even numbers
        print(f"Square of {num} is {num ** 2}")
```

```python

# Hospital Billing System


room_charge_per_day = 22000
doctor_fee = 1500
lab_test_charge = 300

days_stayed = 4
number_of_tests = 3
medicine_charges = 2400

room_charges = room_charge_per_day * days_stayed
lab_charges = lab_test_charge * number_of_tests

total_bill = room_charges + doctor_fee + lab_charges + medicine_charges

if total_bill > 10000:
    discount = total_bill * 0.10
else:
    discount = 0

final_bill = total_bill - discount

print("Total Bill before discount: ₹", total_bill)
print("Discount applied: ₹", discount)
print("Final Bill to be paid: ₹", final_bill)
```
