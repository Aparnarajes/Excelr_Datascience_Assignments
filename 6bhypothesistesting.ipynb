# Hypothesis Testing – Bombay Hospitality Ltd.

This notebook tests whether the actual weekly operating cost of restaurants is higher than the theoretical cost model:

W = 1000 + 5X

## Problem Statement

Bombay Hospitality Ltd. uses the cost model:

W = 1000 + 5X

where X = number of units produced per week.

Given:
- Sample size (n) = 25 restaurants
- Sample mean weekly cost = Rs. 3050
- Mean units produced = 600
- Standard deviation of units = 25
- σ for cost = 5 × 25 = 125
- Significance level α = 0.05

We test whether the actual weekly cost is higher than the model.

## Hypotheses

H₀: μ = 4000 (The cost model is correct)  
H₁: μ > 4000 (Actual weekly cost is higher)

This is a **right-tailed Z-test**.

```python
import math
from scipy.stats import norm
```

```python
# Given values
sample_mean = 3050
mu = 1000 + 5 * 600   # theoretical mean
sigma = 5 * 25        # population standard deviation
n = 25
alpha = 0.05

mu, sigma
```

## Test Statistic

Z = (x̄ − μ) / (σ / √n)

```python
standard_error = sigma / math.sqrt(n)
z = (sample_mean - mu) / standard_error

standard_error, z
```

```python
standard_error = sigma / math.sqrt(n)
z = (sample_mean - mu) / standard_error

standard_error, z
```

## Critical Value

For a right-tailed test at α = 0.05:

Zcritical = 1.645

```python
critical_z = norm.ppf(1 - alpha)
critical_z
```

## Decision Rule

Reject H₀ if Z > 1.645  
Otherwise, fail to reject H₀

```python
if z > critical_z:
    decision = "Reject H0"
else:
    decision = "Fail to Reject H0"

decision
```

## Conclusion

The calculated Z value is **-38**, which is far less than the critical value **1.645**.

Therefore, we fail to reject the null hypothesis.

There is **no statistical evidence** to support the restaurant owners’ claim that the weekly operating costs are higher than the theoretical model.  
In fact, the observed costs are significantly lower than the expected cost.
